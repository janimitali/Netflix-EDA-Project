## Problem Statement:

Netflix has a vast catalog of content. But what makes content popular? Are there directors or genres that dominate? What ratings are common? This project uses EDA to analyze Netflix’s content and extract insights for platform strategy, viewer preferences, and content acquisition.


```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head()
```

```python
#Checking for nulls
df.isnull().sum()
#Converting date formats

#Removing duplicates
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df.head()
```

```python
df['country'] = df['country'].fillna('unknown')
df.head()
```

```python
df['director'] = df['director'].fillna('unknown director')
df.head()
```

```python
df['cast'] = df['cast'].fillna('unknown cast')
df.head()
```

```python
df = df.dropna(subset=['date_added', 'rating', 'duration'])
df.isnull().sum()
```

```python
df.drop_duplicates(inplace =True)
df.info()
```

```python
df['title'] = df['title'].str.strip()
df['director'] = df['director'].str.strip()
df['cast'] = df['cast'].str.strip()
df['listed_in'] = df['listed_in'].str.strip()
df.head(5)
```

## EDA

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Content Breakdown
plt.figure(figsize=(6,4)) #size of the chart
sns.countplot(data=df, x='type',palette ='Set2') #chart type and color

plt.title('distribution of movies and TV shows on Netflix')
plt.xlabel('Type')
plt.ylabel('count')
plt.show()
```

```python
#calculate the % of movies and tv show

Total = df['type'].value_counts().sum()
movie_percentage =(df['type'].value_counts()['Movie']/Total)*100
TVshow_percentage =(df['type'].value_counts()['TV Show']/Total)*100

#string literal we use f"typw:{variable:2f}%"
print(f"Movies: {movie_percentage:.2f}%")
print(f"Tv shows: {TVshow_percentage:.2f}%")
```

```python
# Top Directors
Top_directors = df['director'].value_counts().head(5)
print(Top_directors)
```

```python
#top director with highest rating films or tv 
Top_directors_withRating = df.groupby('director').agg(
    num_titles =('title', 'count'),
    highest_rating=('rating','max')
).sort_values(by='num_titles', ascending=False).head(10)
print(Top_directors_withRating)
```

## Genre Analysis (EDA on listed_in column)

```python
# Split the genre string and explode
df['genre'] = df['listed_in'].str.split(', ')
df_exploded = df.explode('genre')

# Top 10 genres
top_genres = df_exploded['genre'].value_counts().head(10)
print(top_genres)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='Set3')
plt.title('Top 10 Most Common Netflix Genres')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

##  Content Rating Trends Over Time

```python
rating_year = df.groupby(['release_year', 'rating']).size().reset_index(name='count')

plt.figure(figsize=(12,6))
sns.lineplot(data=rating_year, x='release_year', y='count', hue='rating')
plt.title('Content Ratings Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Rating')
plt.show()
```

## Country analysis

Insight : The .explode() function is used when a single column contains lists (or arrays), and you want to turn each element of those lists into a separate row—while keeping the other column values the same.

```python
# Split and explode countries
df['country_split'] = df['country'].str.split(', ')
df_country = df.explode('country_split')

# Top countries
top_countries = df_country['country_split'].value_counts().head(10)
print(top_countries)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Mapping TV ratings to numeric values
rating_map = {
    'TV-Y': 0,
    'TV-Y7': 1,
    'TV-G': 2,
    'TV-PG': 3,
    'TV-14': 4,
    'TV-MA': 5,
    'NR': 0  # 'NR' (Not Rated) can be treated as 0 or omitted
}

# Replace the TV ratings with numeric values
df['rating_numeric'] = df['rating'].map(rating_map)

# Check for any 'NaN' values that couldn't be mapped and handle them
df['rating_numeric'] = df['rating_numeric'].fillna(0)  # Fill missing ratings with 0 or any other default value

# Verify
print(df[['rating', 'rating_numeric']].head())
```

```python
# Convert duration to numeric for Movies
df['duration_numeric'] = df['duration'].str.extract('(\d+)').astype(float)

# Only for Movies
df_corr = df[df['type'] == 'Movie'][['rating_numeric', 'release_year', 'duration_numeric']]
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Movies)')
plt.show()
```

## Analysis

📊 Interpreting the Heatmap:
1. rating_numeric vs release_year → 0.22
✅ Weak positive correlation
As years progress, content tends to get slightly higher (more mature) ratings. But it’s not a strong relationship.

2. rating_numeric vs duration_numeric → 0.016
⚪ No meaningful correlation
Movie rating has almost no relation to how long the movie is.

3. release_year vs duration_numeric → ~0 or slightly negative
🔴 Again, very weak or no correlation
The release year doesn’t affect movie length significantly.
